```python
import sqlite3 as sql

def createDB():
    conn = sql.connect('streamers.db') #Objeto conceccion. Este seria el 'mensajero' o el tunel de coneccion
    conn.commit()
    conn.close()


def createTable():
    conn = sql.connect('streamers.db')
    cursor = conn.cursor() #Objeto de la coneccion. Es como el seleccionador o bien como el mismo nombre dice, el cursor
    cursor.execute(                     #Doc string es con """ """
        """CREATE TABLE streamers (
            name text, 
            followers integer, 
            subs integer
        )"""
    )
    conn.commit()
    conn.close()

def insertRow(nombre, followers, subs):
    conn = sql.connect('streamers.db')
    cursor = conn.cursor() #Objeto de la coneccion. Es como el seleccionador o bien como el mismo nombre dice, el cursor
    instruccion = f"INSERT INTO streamers VALUES ('{nombre}', {followers}, {subs})" #Los valores de tipo string, van entre comillas
    cursor.execute(instruccion)
    conn.commit()
    conn.close()

def readRows():
    conn = sql.connect('streamers.db')
    cursor = conn.cursor() #Objeto de la coneccion. Es como el seleccionador o bien como el mismo nombre dice, el cursor
    instruccion = f"SELECT * FROM streamers"
    cursor.execute(instruccion)
    datos = cursor.fetchall() #Guarda todos los datos seleccionados en 'datos'. 'fetchall()' devuelve una lista con todos los datos
    conn.commit()
    conn.close()
    print(datos)

def insertRows(streamerlist): #Insertar varias filas
    conn = sql.connect('streamers.db')
    cursor = conn.cursor() 
    instruccion = f"INSERT INTO streamers VALUES (?,?, ?)"
    cursor.executemany(instruccion, streamerlist)
    conn.commit()
    conn.close()

def readOrderRead(field):
    conn = sql.connect('streamers.db')
    cursor = conn.cursor()
    instruccion = f"SELECT * FROM streamers ORDER BY {field} DESC"
    cursor.execute(instruccion)
    datos = cursor.fetchall() 
    conn.commit()
    conn.close()
    print(datos)

def search(name):
    conn = sql.connect('streamers.db')
    cursor = conn.cursor()
    instruccion = f"SELECT * FROM streamers WHERE name = '{name}'" #En vez de '=' puedo usar el 'LIKE' que es como poner que sea parecido. Sirve para que sea indiferente las mayusculas y/o ortografias
    cursor.execute(instruccion)
    datos = cursor.fetchall() 
    conn.commit()
    conn.close()
    print(datos)

def updateFields():
    conn = sql.connect('streamers.db')
    cursor = conn.cursor()
    instruccion = f"UPDATE streamers SET followers = 1200000 WHERE name LIKE 'elxokas'" 
    cursor.execute(instruccion) 
    
    conn.commit()
    conn.close()

def deleteRow(): 
    conn = sql.connect('streamers.db')
    cursor = conn.cursor()
    instruccion = f"DELETE FROM streamers WHERE name LIKE 'auronplay'" 
    cursor.execute(instruccion) 
    
    conn.commit()
    conn.close()


if __name__ == '__main__':
    createDB()
    createTable()
    insertRow('Ibai', 7000000, 25000)
    insertRow('AlexElCapo', 800000, 10000)
    readRows()
    streamers = {
        ('ElXokas', 1000000, 9500), 
        ('Cristinini', 300000, 5500), 
        ('Auronplay', 8000000, 20000)
    }
    insertRows(streamers)
    readOrderRead('subs')
    search('Auronplay')
    updateFields()
    deleteRow()
    pass
```
